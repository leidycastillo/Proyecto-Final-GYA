```python
def encontrar_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

print(encontrar_outliers(df, 'CANTIDAD').value_counts().sum())
print(encontrar_outliers(df, 'VALOR_PACTADO').value_counts().sum())
print(encontrar_outliers(df, 'VALOR_PAGADO').value_counts().sum())
print(encontrar_outliers(df, 'HORAS_VIAJE').value_counts().sum())
print(encontrar_outliers(df, 'HORAS_CARGUE').value_counts().sum())
print(encontrar_outliers(df, 'HORAS_DESCARGUE').value_counts().sum())
print(encontrar_outliers(df, 'HORAS_ESPERA_CARGUE').value_counts().sum())
print(encontrar_outliers(df, 'HORAS_ESPERA_DESCARGUE').value_counts().sum())
```

#ANALISIS EXPLORATORIO DE DATOS

```python
config_count = df['CONFIGURACION'].value_counts()

plt.figure(figsize=(10, 6))
config_count.plot(kind='bar', color='skyblue')
plt.title('Camiones por configuracion')
plt.xlabel('Configuracion')
plt.ylabel('Numero de camiones')
plt.xticks(rotation=45)
plt.show()
```

```python
config_count = df['NATURALEZA'].value_counts()

plt.figure(figsize=(10, 6))
config_count.plot(kind='bar', color='skyblue')
plt.title('Camiones por naturaleza')
plt.xlabel('naturaleza')
plt.ylabel('Numero de camiones')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Day'] = df['FECHA_SALIDA'].dt.day
avg_travel_time_per_day = df.groupby('Day')['HORAS_VIAJE'].mean()

plt.figure(figsize=(12, 6))
avg_travel_time_per_day.plot(kind='line', marker='o', color='orange')
plt.title('Promedio de horas_viaje por día en enero 2023')
plt.xlabel('Día')
plt.ylabel('Promedio tiempo de viaje (horas)')
plt.xticks(avg_travel_time_per_day.index)
plt.grid()
plt.show()
```

Los dias lunes y martes son los dias en los que los envios duran menos

```python
daily_shipments = df.groupby(df['FECHA_SALIDA'].dt.day).size()

plt.figure(figsize=(14, 7))
plt.plot(daily_shipments.index, daily_shipments.values, marker='o', color='blue', linewidth=2, markersize=8)

plt.title('Número de envíos por día en enero 2023', fontsize=16)
plt.xlabel('Día', fontsize=14)
plt.ylabel('Número de envíos', fontsize=14)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xticks(daily_shipments.index)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.ticker as ticker

# Convert CANTIDAD to numeric, if it's not already
df['CANTIDAD'] = pd.to_numeric(df['CANTIDAD'], errors='coerce')

# Set the figure size for correlation
plt.figure(figsize=(10, 6))

# Create the scatter plot
sns.scatterplot(x='CANTIDAD', y='HORAS_VIAJE', data=df, color='purple', alpha=0.6)

# Set the title and labels
plt.title('Correlación entre CANTIDAD y HORAS_VIAJE', fontsize=16)
plt.xlabel('Cantidad de Carga', fontsize=14)
plt.ylabel('Horas de Viaje', fontsize=14)

# Format the x-axis to avoid scientific notation
plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Show the plot
plt.grid()
plt.show()
```

```python
avg_travel_time_by_company = df.groupby('EMPRESA_TRANSPORTE')['HORAS_VIAJE'].mean().sort_values()

top_10_companies = avg_travel_time_by_company.tail(10)

plt.figure(figsize=(12, 6))
top_10_companies.plot(kind='bar', color='lightgreen')
plt.title('Promedio de Horas de Viaje por las 10 Empresas de Transporte con Mayor Promedio', fontsize=16)
plt.xlabel('Empresa de Transporte', fontsize=14)
plt.ylabel('Promedio de Horas de Viaje', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
